```python
ids = open('data.txt').read().splitlines()
```

```python
video_list = []
for id in ids:
  url = "https://imvdb.com/api/v1/video/{}?include=popularity".format(id)

  print(url)
  try:
    JSONContent = requests.get(url).json()
    if 'error' not in JSONContent and JSONContent['directors'] and JSONContent['artists']:
        video_list.append([JSONContent['id'], JSONContent['song_title'], JSONContent['artists'][0]['name'],
                             JSONContent['directors'][0]['entity_name'], JSONContent['popularity']['views_all_time'], JSONContent['year']])
  except ValueError:  # includes simplejson.decoder.JSONDecodeError
    print('Decoding JSON has failed')

```

```python
dataset = pd.DataFrame(video_list)
dataset.sample(6)
```

```python
dataset.columns = ['Id', 'Song name', 'Artist', 'Director', 'Views', 'Year']
dataset.dropna(axis = 0, how = 'any', inplace = True)
dataset.index = pd.RangeIndex(len(dataset.index))
dataset
```

```python
dataset.to_csv("Dataset.csv", index = False)
```
